### regresion logistica

 Qu la diferencia de la regresin lineal?

- Regresin lineal predice un nmero continuo (como temperatura, precio, etc).
- Regresin logstica predice probabilidades y luego las convierte en una clase.

 Para qu sirve?

Se usa cuando quieres responder preguntas como:

- Este correo es spam o no?
- El cliente comprar o no?
- Un paciente tiene una enfermedad o no?





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

df = pd.read_csv(r"C:\Users\steve\OneDrive\Documentos\python\base_datos_2008.csv")
```

```python
df = df.dropna(subset=["ArrDelay"])
df = df.sample(frac=1).head(100000)
Y = df["ArrDelay"] < 30 # True = 1/ False = 0 // Interpretacion: Vuelos sin retraso notable
X = df[["DepDelay"]]

Y
```

```python
X
```

```python
logreg = LogisticRegression()
logreg.fit(X, Y)
Y_pred = logreg.predict(X)
```

```python
np.round(logreg.predict_proba(X), 3)
```

 Qu hace?
Esta instruccin predice las probabilidades de clase usando un modelo de regresin logstica (logreg), y redondea los resultados a 3 decimales.



```python
np.mean(Y_pred == Y)
```

```python
np.mean(Y) # Esto me dice que si me preguntan que el vuelo se retrasara, 
#Podemos responder que no, con un 86.64% de probabilidad
```

```python
confusion_matrix = confusion_matrix(Y, Y_pred)
print(confusion_matrix)
```

 Qu es confusion_matrix?
La matriz de confusin es una tabla que resume el desempeo de un modelo de clasificacin, comparando los valores reales (Y) con los predichos (Y_pred).
